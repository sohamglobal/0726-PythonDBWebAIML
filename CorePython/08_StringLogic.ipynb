reverse the string

```python
s="sohamglobal"
rev=""

for ch in s:
  rev=ch+rev

print(rev)
```

palindrome

```python
s="madam"
s1=s[::-1]

if s==s1:
  print('the string is a palindrome')
else:
  print('not a palindrome')
```

number of vowels in a string

```python
s="technology is future"
vowels="aeiouAEIOU"
count=0

for ch in s:
  if ch in vowels:
    count+=1

print(f"Number of vowels in the string are {count}")
```

Anagram

```python
s1="secure"
s2="rescue"
print(sorted(s1))
print(sorted(s2))

if sorted(s1)==sorted(s2):
  print('anagram')
else:
  print('not anagram')
```

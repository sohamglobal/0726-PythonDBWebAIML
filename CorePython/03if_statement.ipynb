```python
amt=float(input('Enter purchase amount : '))
if amt>10000:
  disc=amt*13/100
  print('you have got 13% discount')
else:
  disc=amt*9/100
  print('you have got 9% discount')

print(f'discount is {disc}')
final_bill=amt-disc
print(f'final payable bill is {final_bill}')
```

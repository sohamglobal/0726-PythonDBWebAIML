```python
skills=[]
cho=''

while cho.lower()!='no':
  sk=input('Enter a skill : ')
  skills.append(sk)
  cho=input('Do you have any more skill? (yes/no) ')

print(skills)
```

```python
skills=[]
cho=''

while cho.lower()!='no':
  sk=input('Enter a skill : ')
  if sk not in skills:
    skills.append(sk)
  cho=input('Do you have any more skill? (yes/no) ')

print(skills)
```

**1. Numeric -> int, float, complex**

```python
a=9
print(type(a))
b=int(input('Enter a number : '))
print(type(b))
```

```python
x=45.79
print(type(x))
y=float(input('Enter a number : '))
print(type(y))
```

```python
# complex number is combination of REAL+IMAGINARY part
z1=3+4j
z2=complex(6,9)
z3=z1+z2
print(z3)
print(z3.real)
print(z3.imag)
```

**2. Sequence -> string, list, tuple, range**

```python
nm='sohamglobal'
print(type(nm))
countries=['england','germany','netherlands','france','japan']
print(type(countries))
languages=('english','urdu','french','spanish')
print(type(languages))
n=range(5,11)
print(type(n))
```

**3. Set -> set, frozenset**

```python
names={'praffull','sharayu','soham','shailaja'}
print(type(names))
```

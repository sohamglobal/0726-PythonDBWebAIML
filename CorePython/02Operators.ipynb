**Arithmetic operators**

```python
a=10
b=3
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a%b)
print(b**2)
```

Comparison/Relational operators

```python
x=9
y=13

print(x==y)
print(x!=y)
print(x<y)
print(x>y)
print(x<=y)
print(x>=y)
```

Assignment operators

```python
x=5
x+=2
x-=1
x*=3
x/=2
print(x)
```

Logical Operators

```python
age=25
salary=30000

print(age>18 and salary>45000)
print(age>18 or salary>45000)
print(not(age>18))
```
